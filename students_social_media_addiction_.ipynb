```python
import pandas as pd
import plotly.express as px
```

```python
tabela = pd.read_csv('/content/drive/MyDrive/csv/students_social_media_addiction.csv', encoding='latin1')
tabela = tabela.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'Unnamed: 0.2'], errors='ignore')

display(tabela)
```

```python

tabela = tabela.rename(columns={
    'Student_ID': 'ID_estudante',
    'Age': 'Idade',
    'Gender': 'Genero',
    'Academic_Level': 'Nivel_Academico',
    'Country':'Pais',
    'Avg_Daily_Usage_Hours': 'Horas_Media_de_Uso_Diario',
    'Most_Used_Platform': 'Plataforma_mais_usada',
    'Affects_Academic_Performance': 'Desempenho_Academico_Afetado',
    'Sleep_Hours_Per_Night':'Horas_de_sono_por_noite',
    'Mental_Health_Score':'Pontuacao de Saude Mental',
    'Relationship_Status':'Status_Relacionamento',
    'Conflicts_Over_Social_Media':'Conflitos_nas_Midias_Sociais',
    'Addicted_Score':'Pontuacao_viciado'
})
tabela.head()
```

```python
tabela.dropna()
display(tabela.info())
```

```python
qts_pla = tabela['Plataforma_mais_usada'].value_counts()
qts_genero = tabela['Genero'].value_counts()
qts_nivel = tabela['Nivel_Academico'].value_counts()
qts_status = tabela['Status_Relacionamento'].value_counts()


print(f'Quantidade de Plataforma {qts_pla}')
print('--------------------------------')
print(f'Quantidade de GÃªnero {qts_genero}')
print('--------------------------------')
print(f'Quantidade de NÃ­vel {qts_nivel}')
print('--------------------------------')
print(f'Quantidade de Status {qts_status}')

```

```python
categoria = tabela.select_dtypes(include='object')
for col in categoria:
    print(f'Coluna: {col}')
    print(tabela[col].value_counts())
    print('--------------------------------')
```

```python
tabela['Genero'] = tabela['Genero'].replace({
    'Female': 'Feminino',
    'Male': 'Masculino'
})
```

```python
tabela['Nivel_Academico'] = tabela['Nivel_Academico'].replace({
      'Undergraduate': 'GraduaÃ§Ã£o',
      'Graduate': 'Diplomado',
      'High School': 'Ensino mÃ©dio'


})
```

```python
tabela['Status_Relacionamento'] = tabela['Status_Relacionamento'].replace({
    'In Relationship':'Em relacionamento',
    'Single': 'Solteiro',
    'Complicated': 'Complicado'

})
```

```python
# GrÃ¡fico1 de proporÃ§Ã£o (pizza)
# Qual plataforma Ã© mais popular?
grafico1 = px.pie(
    tabela,
    names='Plataforma_mais_usada',
    title='Plataforma Mais Usada pelos Estudantes'
)
grafico1.show()
```

```python
# MÃ©dia de horas de uso por plataforma
media_por_plataforma = tabela.groupby('Plataforma_mais_usada')['Horas_Media_de_Uso_Diario'].mean().reset_index()
```

```python

grafico2 = px.bar(
    media_por_plataforma,
    x='Plataforma_mais_usada',
    y='Horas_Media_de_Uso_Diario',
    text_auto=True,
    title='MÃ©dia de Horas de Uso DiÃ¡rio por Plataforma'
)

grafico2.show()
```

```python
# DispersÃ£o
# RelaÃ§Ã£o entre uso de mÃ­dias e horas de sono
grafico4 = px.scatter(
    tabela,
    x='Horas_Media_de_Uso_Diario',
    y='Horas_de_sono_por_noite',
    color='Plataforma_mais_usada',
    size='Horas_Media_de_Uso_Diario',
    title='CorrelaÃ§Ã£o entre Uso de Midias e Sono'
)
grafico4.show()
```

```python
grafico5 = px.histogram(
    tabela,
    x='Desempenho_Academico_Afetado',
    color= 'Plataforma_mais_usada',
    barmode= 'group',
    title='Impacto no Desempenho AcadÃªmico por Plataforma'
)
grafico5.show()
```

```python
for grafico in [grafico1, grafico2, grafico4, grafico5]:
    grafico.show()
```
