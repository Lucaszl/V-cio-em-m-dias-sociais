```python
import pandas as pd
import plotly.express as px
```

```python
tabela = pd.read_csv('/content/drive/MyDrive/csv/Vício_em_mídias_sociais_para_estudantes.csv', encoding='latin1')
tabela = tabela.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'Unnamed: 0.2'], errors='ignore')
print(tabela.columns)
tabela.head()
```

```python

tabela = tabela.rename(columns={
    'Student_ID': 'ID_estudante',
    'Age': 'Idade',
    'Gender': 'Gênero',
    'Academic_Level': 'Nível_Acadêmico',
    'Country':'País',
    'Avg_Daily_Usage_Hours': 'Horas_Média_de_Uso_Diário',
    'Most_Used_Platform': 'Plataforma_mais_usada',
    'Affects_Academic_Performance': 'Desempenho_Acadêmico_Afetado',
    'Sleep_Hours_Per_Night':'Horas_de_sono_por_noite',
    'Mental_Health_Score':'Pontuação de Saúde Mental',
    'Relationship_Status':'Status_Relacionamento',
    'Conflicts_Over_Social_Media':'Conflitos_nas_Mídias_Sociais',
    'Addicted_Score':'Pontuação_viciado'
})
tabela.head()
```

```python
tabela.info()
```

```python
tabela.dropna()
```

```python
qts_pla = tabela['Plataforma_mais_usada'].value_counts()
qts_genero = tabela['Gênero'].value_counts()
qts_nivel = tabela['Nível_Acadêmico'].value_counts()
qts_status = tabela['Status_Relacionamento'].value_counts()


print(f'Quantidade de Plataforma {qts_pla}')
print('--------------------------------')
print(f'Quantidade de Gênero {qts_genero}')
print('--------------------------------')
print(f'Quantidade de Nível {qts_nivel}')
print('--------------------------------')
print(f'Quantidade de Status {qts_status}')

```

```python
categoria = tabela.select_dtypes(include='object')
for col in categoria:
    print(f'Coluna: {col}')
    print(tabela[col].value_counts())
    print('--------------------------------')
```

```python
tabela['Gênero'] = tabela['Gênero'].replace({
    'Female': 'Feminino',
    'Male': 'Masculino'
})
```

```python
tabela['Nível_Acadêmico'] = tabela['Nível_Acadêmico'].replace({
      'Undergraduate': 'Graduação',
      'Graduate': 'Diplomado',
      'High School': 'Ensino médio'


})
```

```python
tabela['Status_Relacionamento'] = tabela['Status_Relacionamento'].replace({
    'In Relationship':'Em relacionamento',
    'Single': 'Solteiro',
    'Complicated	': 'Complicado'

})
```

```python
display(tabela)
```

```python
# Gráfico1 de proporção (pizza)
# Qual plataforma é mais popular?
grafico1 = px.pie(
    tabela,
    names='Plataforma_mais_usada',
    title='Plataforma Mais Usada pelos Estudantes'
)
grafico1.show()
# Podemos observer qual é plataforma mais usada - Instagram 35.3%
```

```python
# Média de horas de uso por plataforma
media_por_plataforma = tabela.groupby('Plataforma_mais_usada')['Horas_Média_de_Uso_Diário'].mean().reset_index()
```

```python

grafico2 = px.bar(
    media_por_plataforma,
    x='Plataforma_mais_usada',
    y='Horas_Média_de_Uso_Diário',
    text_auto=True,
    title='Média de Horas de Uso Diário por Plataforma'
)

grafico2.show()
```

```python
# Dispersão
# Relação entre uso de mídias e horas de sono
grafico4 = px.scatter(
    tabela,
    x='Horas_Média_de_Uso_Diário',
    y='Horas_de_sono_por_noite',
    color='Plataforma_mais_usada',
    size='Horas_Média_de_Uso_Diário',
    title='Correlação entre Uso de Mídias e Sono'
)
grafico4.show()
```

```python
grafico5 = px.histogram(
    tabela,
    x='Desempenho_Acadêmico_Afetado',
    color= 'Plataforma_mais_usada',
    barmode= 'group',
    title='Impacto no Desempenho Acadêmico por Plataforma'
)
grafico5.show()
```
